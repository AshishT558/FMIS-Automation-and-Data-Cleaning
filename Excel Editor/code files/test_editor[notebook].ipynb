```python
import pandas as pd
from datetime import datetime
import xlsxwriter
import os
```

```python
dir_path = "C:\\Users\\athomas2\\Desktop\\Selenium scripts\\FMIS Automation and Data Cleaning\\Excel Editor\\"
end_path = "C:\\Users\\athomas2\\Desktop\\Selenium scripts\\FMIS Automation and Data Cleaning\\Excel Editor\\Edited Files\\"
```

```python
list_excel = []
for file in os.listdir(dir_path):
    if file.endswith(".xlsx"):
        list_excel.append(file)
list_excel
```

```python
name = list_excel[0]
sheet = pd.read_excel(dir_path + name, header=None)
sheet.head()
```

**Remove first row and reassign column names**

```python
#remove the first row
sheet = sheet.iloc[1:]

#set columns
sheet.columns = sheet.iloc[0]

#remove repeat row
sheet = sheet.iloc[1:]
sheet.head()
```

**Add Query Date Column**

```python
#get index where column should be added
column_count = sheet.columns.size

# get current datetime and
# format into mm/dd/yyyy hh:mm:ss AM/PM
now = datetime.now()
formatted_date = now.strftime("%m/%d/%Y %H:%M:%S %p")

#add new column with all entries being the datetime
sheet.insert(column_count, 'Query-Date', formatted_date)
sheet
```

**Rename and export file**

```python
#get index of last '_' in the file_name - this indicates where the number starts
index = name.rindex('_')
file_name = name[:index] + '.xlsx'
```

```python
writer = pd.ExcelWriter(dir_path +"Edited Files\\" + file_name, engine='xlsxwriter')
```

```python
#export
sheet.to_excel(writer, sheet_name="sheet1", index=False)
```

```python
#get the xlsx writer workbook and worksheet objects
workbook = writer.book
worksheet = writer.sheets["sheet1"]
```

```python
#adjust the column widths based on the content
for i, col in enumerate(sheet.columns):
    width = max(sheet[col].apply(lambda x: len(str(x))).max(), len(col))
    worksheet.set_column(i, i, width)
writer.close()
```
